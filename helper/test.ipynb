```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle

# Recreate and refit vectorizer on same training data
vectorizer = TfidfVectorizer(max_features=1, stop_words='english')
vectorizer.fit(train_data)

# Save it so you never lose it again
pickle.dump(vectorizer, open('vectorizer.sav', 'wb'))
```

```python
from google.oauth2.credentials import Credentials
from googleapiclient.discovery import build

# Load credentials generated from OAuth flow
creds = Credentials.from_authorized_user_file('credentials.json', ['https://www.googleapis.com/auth/business.manage'])
service = build('mybusinessbusinessinformation', 'v1', credentials=creds)

# Your business location ID
location_id = 'locations/09069961256292374683'

# Fetch reviews using the correct endpoint
try:
    response = service.locations().get(name=location_id, readMask='reviews').execute()
    
    reviews = response.get('reviews', [])
    print(f"Total reviews found: {len(reviews)}\n")
    
    for i, review in enumerate(reviews, 1):
        print(f"Review {i}:")
        print("Reviewer:", review.get('reviewer', {}).get('displayName', 'Anonymous'))
        print("Rating:", review.get('starRating', 'N/A'))
        print("Comment:", review.get('comment', ''))
        print("Date:", review.get('publishTime', 'N/A'))
        print("-" * 50)
        
except Exception as e:
    print(f"Error: {e}")
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the Google reviews page
url = "https://www.google.com/maps/place/The+papa's+business/@28.6282267,77.1454743,17z/data=!4m8!3m7!1s0x390d03d9a8ed3069:0x451aeec2246de6bf!8m2!3d28.6282267!4d77.1454743!9m1!1b1!16s%2Fg%2F11ylv8lxzt?hl=en&entry=ttu&g_ep=EgoyMDI1MTAyOS4yIKXMDSoASAFQAw%3D%3D"
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# Find review containers using identified classes
review_containers = soup.find_all('span', class_='m6QErb DxyBCb kA9KIf dS8AEf XiKgde ')

for container in review_containers:
    review_text = container.get_text()
    print(f'Review: {review_text}')
```
